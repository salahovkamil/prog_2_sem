```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
month = "02"
year = "2024"
outcome_data = pd.read_excel(f"/home/salahovkamil/outcome_{month}.{year}.xlsx")
outcome_data
```

```python
outcome_data["День"] = [int(x.split()[0]) for x in outcome_data["Дата"]]
outcome_data
```

```python
fig, ax = plt.subplots(constrained_layout=True)
sns.lineplot(
    data=outcome_data,
    x="День",
    y="Сумма",
    hue="Категория",
    ax=ax
)
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(constrained_layout=True)
sns.barplot(
    data=outcome_data,
    x="Сумма",
    y="Категория",
    orient = "h",
    estimator="sum",
    errorbar=None,
    ax=ax
)
plt.title(f'{month}.{year}')
plt.show()
```
